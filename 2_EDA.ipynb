```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/titanic_clean.csv')
```

```python
df.sample(4)
```

```python
df['Survived'].value_counts()
```

```python
plt.figure(figsize=(12, 8))

sns.barplot(x = df['Survived'].value_counts().index, y = df['Survived'].value_counts())
plt.title('Value Counts of Survived')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='Pclass', hue='Survived')

# Personalizar etiquetas y ttulo
plt.title('Tasa de Supervivencia por Clase')
plt.xlabel('Clase')
plt.ylabel('Total')
plt.legend(title='Supervivencia', labels=['No sobrevivi', 'Sobrevivi'])

# Mostrar la grfica
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='Sex', hue='Survived')

# Personalizar etiquetas y ttulo
plt.title('Tasa de Supervivencia por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Total')
plt.legend(title='Supervivencia', labels=['No sobrevivi', 'Sobrevivi'])

# Mostrar la grfica
plt.show()
```

```python
plt.figure(figsize = (12, 8))

# colores para cada una de las clases
color = ['#240750', '#344C64']

sns.countplot(df, x = 'Embarked', hue = 'Survived', palette=color) 
plt.legend(title = 'Supervivencia', labels = ['No sobrevivi', 'S sobrevivi'])
plt.show()
```

```python
plt.figure(figsize = (12, 8))

sns.histplot(df, x = 'Age', kde = True)
plt.title('Distribucin de Edades')
plt.axvline(df['Age'].mean(), color = 'red')
plt.axvline(df['Age'].median(), color = 'green')
plt.axvline(df['Age'].mode()[0], color = 'black')
plt.show()

# Print the results
print(f"Media: {df['Age'].mean()}")
print(f"Mediana: {df['Age'].median()}")
print(f"Moda: {df['Age'].mode()[0]}")
```

```python
plt.figure(figsize = (12, 8))
sns.heatmap(df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr(), 
            annot=True, 
            cbar=True,  # Activar la barra de colores
            cmap='coolwarm',  # Cambiar la paleta de colores
            cbar_kws={'label': 'Nivel de correlacin'})  # Etiqueta para la barra de colores
plt.title('Correlacin de variables numricas')
plt.show()
```

```python
#Agrupa las edades en categoras (Infante, Nio, Adolescente, Joven Adulto, Adulto, Mayor) y crea una grfica que muestre la tasa de supervivencia por grupo de edad.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear categoras de edades
bins = [0, 2, 12, 18, 35, 60, 100]
labels = ['Infante', 'Nio', 'Adolescente', 'Joven Adulto', 'Adulto', 'Mayor']

df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Crear grfica de tasa de supervivencia por grupo de edad
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='AgeGroup', y='Survived')

plt.title('Tasa de Supervivencia por Grupo de Edad')
plt.xlabel('Grupo de Edad')
plt.ylabel('Tasa de Supervivencia')

plt.show()
```

```python
#Crea una grfica que muestre el conteo de pasajeros en cada clase (Pclass).
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Pclass')

plt.title('Conteo de Pasajeros por Clase')
plt.xlabel('Clase de Pasajero (Pclass)')
plt.ylabel('Cantidad')

plt.show()
```

```python
#Crea una grfica que muestre la distribucin de gnero dentro de cada clase de pasajero (Pclass).
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='Pclass', hue='Sex')

plt.title('Distribucin de Gnero por Clase de Pasajero (Pclass)')
plt.xlabel('Clase de Pasajero')
plt.ylabel('Cantidad de Pasajeros')

plt.legend(title='Gnero')
plt.show()
```

```python
#Crea una grfica que muestre el conteo de pasajeros por gnero (Sex).
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Sex')
plt.title('Conteo de Pasajeros por Gnero')
plt.xlabel('Gnero')
plt.ylabel('Cantidad')
plt.show()
```

```python
#Encuentra y muestra los valores mximo y mnimo de la columna de edad.
edad_min = df['Age'].min()
edad_max = df['Age'].max()

print("Edad mnima:", edad_min)
print("Edad mxima:", edad_max)
```

```python
#Crea una grfica que muestre la distribucin de edades, incluyendo lneas verticales que indiquen la media, mediana y moda de las edades.
from statistics import mode
media = df['Age'].mean()
mediana = df['Age'].median()

# Para calcular la moda ignorando valores nulos
moda = df['Age'].dropna().mode()[0]

plt.figure(figsize=(12, 6))
sns.histplot(df['Age'], kde=True, bins=30)

plt.axvline(media, color='red', linestyle='--', label=f'Media: {media:.2f}')
plt.axvline(mediana, color='green', linestyle='--', label=f'Mediana: {mediana:.2f}')
plt.axvline(moda, color='blue', linestyle='--', label=f'Moda: {moda:.2f}')

plt.title("Distribucin de Edades con Media, Mediana y Moda")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.legend()

# Mostrar grfica
plt.show()
```

```python
#Crea un diagrama de caja (boxplot) que muestre los outliers en la columna de edad.
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Age'])

# Ttulo y etiquetas
plt.title('Diagrama de Caja de la Edad (con outliers)')
plt.xlabel('Edad')

# Mostrar grfica
plt.show()
```

```python
#Encuentra y muestra los valores mximo y mnimo de la columna de tarifa (Fare).
max_fare = df['Fare'].max()
min_fare = df['Fare'].min()

print("Tarifa mxima (Fare):", max_fare)
print("Tarifa mnima (Fare):", min_fare)
```

```python
#Crea una grfica que muestre la distribucin de tarifas, incluyendo lneas verticales que indiquen la media, mediana y moda de las tarifas
fare_mean = df['Fare'].mean()
fare_median = df['Fare'].median()
fare_mode = df['Fare'].mode()[0]

plt.figure(figsize=(12, 6))
sns.histplot(df['Fare'], kde=True)

plt.axvline(fare_mean, color='red', linestyle='--', label=f'Media: {fare_mean:.2f}')
plt.axvline(fare_median, color='green', linestyle='--', label=f'Mediana: {fare_median:.2f}')
plt.axvline(fare_mode, color='blue', linestyle='--', label=f'Moda: {fare_mode:.2f}')

plt.title('Distribucin de Tarifas (Fare) con Media, Mediana y Moda')
plt.xlabel('Tarifa')
plt.ylabel('Frecuencia')
plt.legend()

plt.show()
```

```python
#Crea un diagrama de caja (boxplot) que muestre los outliers en la columna de tarifa (Fare).
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['Fare'])

plt.title('Diagrama de Caja de la Tarifa (Fare)')
plt.xlabel('Tarifa')
plt.show()
```

```python
#Crea una grfica que muestre el conteo de pasajeros por nmero de hermanos/esposos a bordo (SibSp).
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='SibSp')

plt.title('Conteo de Pasajeros por Nmero de Hermanos/Esposos a Bordo (SibSp)')
plt.xlabel('Nmero de Hermanos/Esposos (SibSp)')
plt.ylabel('Cantidad de Pasajeros')

plt.show()
```

```python
#Calcula y muestra el porcentaje de supervivencia para cada valor de SibSp.
porcentaje_supervivencia = df.groupby('SibSp')['Survived'].mean() * 100

print("Porcentaje de Supervivencia por nmero de Hermanos/Esposos a bordo (SibSp):")
print(porcentaje_supervivencia)
```

```python
#Crea un mapa de calor (heatmap) que muestre la matriz de correlacin de las caractersticas numricas en el dataset.
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Crear el heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(
    numeric_df.corr(),
    annot=True,        # Mostrar valores dentro del mapa
    cmap='coolwarm',   # Paleta de colores
    cbar=True,         # Mostrar barra de color
    linewidths=0.5     # Lneas entre celdas
)

plt.title("Matriz de Correlacin de Variables Numricas")
plt.show()
```
